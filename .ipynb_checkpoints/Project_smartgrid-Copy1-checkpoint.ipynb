```python
# List of all numerical and continuous columns
numerical_columns = [
    'Temperature (?C)', 'Humidity (%)', 'Rainfall Chances (%)', 'Wind Speed (km/h)',
    'Fan Power Consumed (kWh)', 'Light Power Consumed (kWh)', 'Mixer Power Consumed (kWh)',
    'Washing Machine Power Consumed (kWh)', 'Phone Charging Power Consumed (kWh)',
    'UPS Power Consumed (kWh)', 'Grinder Power Consumed (kWh)', 'AC Power Consumed (kWh)',
    'Heater Power Consumed (kWh)', 'Fridge Power Consumed (kWh)', 'TV Power Consumed (kWh)',
    'Total Power Consumed (kWh)', 'Required Power Supply (kW)', 'Current Power Supply (kW)'
]

# Function to detect and remove outliers using IQR
def remove_outliers_iqr(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)  # First quartile (25%)
        Q3 = df[col].quantile(0.75)  # Third quartile (75%)
        IQR = Q3 - Q1               # Interquartile range
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Print summary for outliers
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        print(f"Column: {col}")
        print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
        print(f"Number of Outliers: {len(outliers)}\n")
        
        # Remove outliers
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    
    return df

# Apply the function to remove outliers
df_cleaned = remove_outliers_iqr(data, numerical_columns)

# Save the cleaned dataset
df_cleaned.to_csv('cleaned_dataset.csv', index=False)
print("Outliers removed. Cleaned dataset saved as 'cleaned_dataset.csv'.")
```

```python
# Normalize data before model fitting
# it will boost the performance( in neural networks) + transform
from sklearn.preprocessing import MinMaxScaler
# scale of the output and input inthe range 0-1 to match the scale of the layer of LSTM
scaler = MinMaxScaler(feature_range = (0,1)) 
# reshape: convert the univariate 1D array into 2D
y = scaler.fit_transform(np.array(y).reshape(-1,1))
print("Normalizing data before model fitting")
print(y[:10])
```

```python
training_size = int(len(y)*0.80)
test_size = len(y)- training_size
val_size = int(training_size*0.20)
train_data , test_data , val_data = y[0:training_size-val_size,:] , y[training_size:len(y),:1], y[len(y)-test_size-val_size:len(y)-test_size,:1]
```

```python
# building input variable
def create_dataset(dataset, time_step = 1):
  dataX, dataY = [] , []
  for i in range(len(dataset)-time_step-1):
    a = dataset[i:(i+time_step),0]
    dataX.append(a)
    dataY.append(dataset[i + time_step,0])
  return np.array(dataX), np.array(dataY)
```

```python
time_step = 100 
X_train, y_train = create_dataset(train_data, time_step)
X_test, ytest = create_dataset(test_data, time_step)
X_val, yval = create_dataset(val_data, time_step)
```

```python
# reshape train and input-output pairs
X_train = X_train.reshape(X_train.shape[0],X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1],1)
X_val = X_val.reshape(X_val.shape[0], X_val.shape[1],1)

print("X_train shape: ", X_train.shape)
print("X_test shape: ",X_test.shape)
print("X_val shape: ",X_val.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.callbacks import EarlyStopping

model = Sequential()

# Adding the first LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50, return_sequences = True, input_shape = (time_step, 1)))
model.add(Dropout(0.2))

# Adding a second LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50, return_sequences = True))
# model.add(Dropout(0.2))

# # Adding a third LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50, return_sequences = True))
# model.add(Dropout(0.2))

# Adding a fourth LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50))
# model.add(Dropout(0.2))

# Adding the output layer
model.add(Dense(units = 1))

# Compiling the RNN
model.compile(optimizer = 'adam', loss = 'mean_squared_error')

early_stopping = EarlyStopping(
    monitor='val_loss', 
    patience=10,  # Stops training if no improvement after 10 epochs
    restore_best_weights=True
)
```

```python
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, yval),
    epochs=60,
    batch_size=20,
    verbose=1,
    callbacks=[early_stopping]
)
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout

X = data.iloc[:, :19].values  # First 19 columns
y = data.iloc[:, 19:].values  # Last 5 columns

# Initialize scalers
scaler_X = MinMaxScaler(feature_range=(0, 1))
scaler_y = MinMaxScaler(feature_range=(0, 1))

X = scaler_X.fit_transform(X)
y = scaler_y.fit_transform(y)

# Split data
training_size = int(len(X) * 0.80)
val_size = int(training_size * 0.20)
test_size = len(X) - training_size

train_X, val_X, test_X = (
    X[:training_size - val_size],
    X[training_size - val_size:training_size],
    X[training_size:],
)
train_y, val_y, test_y = (
    y[:training_size - val_size],
    y[training_size - val_size:training_size],
    y[training_size:],
)

# Create dataset
def create_dataset(X, y, time_step=1):
    dataX, dataY = [], []
    for i in range(len(X) - time_step):
        dataX.append(X[i:(i + time_step), :])
        dataY.append(y[i + time_step - 1, :])
    return np.array(dataX), np.array(dataY)

time_step = 100
X_train, y_train = create_dataset(train_X, train_y, time_step)
X_val, y_val = create_dataset(val_X, val_y, time_step)
X_test, y_test = create_dataset(test_X, test_y, time_step)

# Reshape input to 3D
X_train = X_train.reshape(X_train.shape[0], time_step, 19)
X_val = X_val.reshape(X_val.shape[0], time_step, 19)
X_test = X_test.reshape(X_test.shape[0], time_step, 19)

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_val shape:", X_val.shape)
print("y_val shape:", y_val.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
from keras.callbacks import EarlyStopping
```

```python
model = Sequential()

model.add(LSTM(units=50, return_sequences=True, input_shape=(time_step, 19)))
model.add(Dropout(0.2))

model.add(LSTM(units=50, return_sequences=True))
model.add(LSTM(units=50))
model.add(Dropout(0.2))

model.add(Dense(units=5))  # Match the number of output columns

# Compile the model
model.compile(optimizer="adam", loss="mean_squared_error")

# Add EarlyStopping to prevent overfitting
early_stopping = EarlyStopping(
    monitor='val_loss', 
    patience=10,  # Stops training if no improvement after 10 epochs
    restore_best_weights=True
)

# Train the model
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=60,
    batch_size=20,
    verbose=1,
    callbacks=[early_stopping]
)

# Evaluate
loss = model.evaluate(X_test, y_test)
print("Test loss:", loss)
```

```python
def series_to_supervised(data, n_in=1, n_out=1):
    """Convert multivariate time series to a supervised dataset while retaining original column count."""
    df = pd.DataFrame(data)
    
    # Create a copy of the original DataFrame
    transformed = df.copy()
    
    # Add lagged values
    for lag in range(1, n_in + 1):
        lagged = df.shift(lag)
        transformed += lagged  # Combine lagged values into the same columns
    
    # Add target values (future steps)
    for lead in range(1, n_out + 1):
        future = df.shift(-lead)
        transformed += future  # Combine future values into the same columns
    
    # Drop rows with NaN values (due to shifting)
    transformed.dropna(inplace=True)
    
    return transformed

# Example Usage
# Assuming `scaled` is a numpy array with shape (50000, 24)
scaled_df = pd.DataFrame(scaled)
reframed = series_to_supervised(scaled_df, n_in=1, n_out=1)

print("Reframed Data shape:", reframed.shape)  # Expected: (50000 - shifts, 24)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import KFold
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

# Define the LSTM-based model
class PowerManagementPredictor(nn.Module):
    def __init__(self, input_size, output_size, hidden_size=64, num_layers=2):
        super(PowerManagementPredictor, self).__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        if lstm_out.dim() == 2:
            output = self.fc(lstm_out)
        else:
            output = self.fc(lstm_out[:, -1, :])  # Use the last time step
        return output
```

```python
import os
from pathlib import Path
# Define input and output sizes
input_size = X_tensor.shape[1]
output_size = y_tensor.shape[1]  # Number of output features

# Instantiate the Power Management Predictor model
model = PowerManagementPredictor(input_size, output_size)
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)
criterion = nn.MSELoss()  # Mean Squared Error Loss for regression

# Training loop with k-fold cross-validation
k_folds = 5
skf = KFold(n_splits=k_folds, shuffle=True, random_state=42)

# Define directory to save model checkpoints
output_dir = Path("model_checkpoints")
output_dir.mkdir(parents=True, exist_ok=True)

# Initialize best model tracking and early stopping counter
best_val_loss = float('inf')
best_model = None
patience = 10
counter = 0

for fold, (train_indices, val_indices) in enumerate(skf.split(X_tensor)):
    print(f'Fold {fold + 1}/{k_folds}')

    X_train, X_val = X_tensor[train_indices], X_tensor[val_indices]
    y_train, y_val = y_tensor[train_indices], y_tensor[val_indices]

    # Reshape inputs for LSTM
    X_train = X_train.unsqueeze(1)  # Add sequence dimension (batch, seq_len, input_size)
    X_val = X_val.unsqueeze(1)
    
    train_dataset = TensorDataset(X_train, y_train)
    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
    val_dataset = TensorDataset(X_val, y_val)
    val_loader = DataLoader(val_dataset, batch_size=32)

    # Initialize model, optimizer, and loss function for each fold
    model = PowerManagementPredictor(input_size, output_size)
    optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)
    criterion = nn.MSELoss()
    
    # Learning rate scheduler
    scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3)

    # Training loop
    train_losses = []
    val_losses = []
    no_improvement_counter = 0
    
    for epoch in range(100):  # Adjust number of epochs as needed
        model.train()
        train_loss = 0.0

        for batch_idx, (inputs, targets) in enumerate(train_loader):
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()

        train_loss /= len(train_loader)
        train_losses.append(train_loss)

        # Validation loop
        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for batch_idx, (inputs, targets) in enumerate(val_loader):
                outputs = model(inputs)
                val_loss += criterion(outputs, targets).item()

        val_loss /= len(val_loader)
        val_losses.append(val_loss)

        print(f'Epoch [{epoch + 1}/100], Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}')

        # Save the best model based on validation loss
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_model = model.state_dict()
            torch.save(best_model, "best_model.pth")
            print(f"Best model updated!")
            counter = 0  # Reset counter if validation loss improves
        else:
            counter += 1  # Increment counter if no improvement

        # Early stopping condition
        if counter >= patience:
            print(f"Early stopping after {epoch + 1} epochs due to no improvement in validation loss.")
            break

# Load the best model after training all folds
model.load_state_dict(best_model)

# Evaluate on the test set
model.eval()
with torch.no_grad():
    test_outputs = model(X_test.unsqueeze(1))  # Reshape test data for LSTM
    test_loss = criterion(test_outputs, y_test).item()
    print(f'Test Loss: {test_loss:.4f}')
    
# Plot training and validation loss
plt.figure(figsize=(10, 6))
plt.plot(train_losses, label="Train Loss")
plt.plot(val_losses, label="Validation Loss", linestyle='--')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss Curves")
plt.legend()
plt.grid(True)
plt.show()
```
